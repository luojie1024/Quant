## A股日线行情

```python
import pandas as pd
import tushare as ts
```

```python
# 设置token
ts.set_token('')

pro = ts.pro_api()
df = pro.daily(ts_code='002436.SZ', start_date='20200101', end_date='20220805')
df.index = pd.to_datetime(df['trade_date'])
df.index.name = None
df.sort_index(inplace=True)
```

```python
df
```

## 画图

```python
df[['close']].plot.line(figsize=(30,14))
```

```python
df[['high','low']].plot.line(figsize=(30,14))
```
